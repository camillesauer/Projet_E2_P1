```python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error


import pandas as pd
import numpy as np
import matplotlib as plt
from typing import List
```

```python
# Loads the California housing Dataset
housing = fetch_openml(name="house_prices", as_frame=True)  # noqa
```

```python
housing = fetch_openml(name="house_prices", as_frame=True) 
data = pd.DataFrame(data=housing.data, columns=housing.feature_names)
data['target']=housing.target
```

Data cleaning

```python
# create dumies
data['Fence'] = data['Fence'].notna()
data['Pool'] = data['PoolArea'] > 0

# create the new variable 'age'
data['Age'] = data.apply(lambda x: x['YrSold']-x['YearBuilt'] if (x['YearBuilt']<x['YearRemodAdd']) 
                                                           else (x['YrSold']-x['YearRemodAdd']), axis=1)

# Delete all data with MSZoning = commercial, agriculture and industrial as these are not residential units
data = data[(data.MSZoning != 'C (all)') & (data.MSZoning != 'I (all)') & (data.MSZoning != 'A (agr)')]

# Delete abnormal sales
data = data[(data.SaleCondition != 'Abnorml')] 
data = data[(data.SaleCondition != 'Family')] 
```

## suppression fence et pool

```python
# Select Features
X = data[['Age','GrLivArea', 'LotFrontage', 'LotArea', 'GarageArea', "TotalBsmtSF", "FullBath", "TotRmsAbvGrd"]]
Y = data[['target']]
```

```python
X.to_csv('clean_X.csv', index=False)
```

```python
X.head()
```

```python
Y.head()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, Y, test_size=.3, random_state=1121218)
```

```python
# transforme les bool√©en en int
X_train=X_train*1
X_valid=X_valid*1
```

```python
numerical_features = X_train.select_dtypes(include='number').columns.tolist()
categorical_features = X_train.select_dtypes(exclude='number').columns.tolist()
```

```python
numeric_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='mean')),
    ('scale', MinMaxScaler())
])

categorical_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('one-hot', OneHotEncoder(handle_unknown='ignore', sparse=False))
])
```

```python
from sklearn.compose import ColumnTransformer

full_processor = ColumnTransformer(transformers=[
    ('number', numeric_pipeline, numerical_features),
    ('category', categorical_pipeline, categorical_features)
])
```

```python
randomf = RandomForestRegressor()

randomf_pipeline = Pipeline(steps=[
    ('preprocess', full_processor),
    ('model', randomf)
])
```

```python
from sklearn.model_selection import GridSearchCV
param_dict = {
    'model__bootstrap': [True],
    'model__max_depth': [80, 90, 100, 110],
    'model__max_features': [2, 3],
    'model__min_samples_leaf': [3, 4, 5],
    'model__min_samples_split': [8, 10, 12],
    'model__n_estimators': [50, 100,150]
}

search = GridSearchCV(randomf_pipeline, param_dict, 
                      cv=5, 
                      scoring='neg_mean_absolute_error')

_ = search.fit(X_train, y_train)

print('Best score:', abs(search.best_score_))

print('Best params:', search.best_params_)
```

```python
feature_importances = search.best_estimator_.steps[1][1].feature_importances_
```

```python
print(len(feature_importances))
```

```python
result=pd.DataFrame()
result["feature_name"]=X_train.columns
result["feature_importances"]=np.round(feature_importances*100,2)
result.head(11).sort_values("feature_importances", ascending = False)
```

```python
y_pred = search.predict(X_valid)
```

```python
print(mean_absolute_error(y_valid, y_pred))
```

```python
Y=y_valid.target.tolist()
```
